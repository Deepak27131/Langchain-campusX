# Dowload own sarver to download and run huggingface models locally

```python

from transformers import pipeline
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline

from dotenv import load_dotenv
load_dotenv()

llm = HuggingFacePipeline(
    repo_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    task="text-generation",
    pipeline_kwargs=dict(
        max_new_tokens= 300,
        temperature= 0.7
    )
)

model = ChatHuggingFace(llm=llm)

response = model.invoke("What about olmaa why it not support mutiple amnd easy.")

print(response)
```

```python
# Install transformers from source - only needed for versions <= v4.34
# pip install git+https://github.com/huggingface/transformers.git
# pip install accelerate

import torch
from transformers import pipeline

pipe = pipeline("text-generation", model="TinyLlama/TinyLlama-1.1B-Chat-v1.0", torch_dtype=torch.bfloat16, device_map="auto")

# We use the tokenizer's chat template to format each message - see https://huggingface.co/docs/transformers/main/en/chat_templating
messages = [
    {
        "role": "system",
        "content": "You are a friendly chatbot who always responds in the style of a pirate",
    },
    {"role": "user", "content": "How many helicopters can a human eat in one sitting?"},
]
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
outputs = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
print(outputs[0]["generated_text"])
# <|system|>
# You are a friendly chatbot who always responds in the style of a pirate.</s>
# <|user|>
# How many helicopters can a human eat in one sitting?</s>
# <|assistant|>
# ...
```

```python
# other method use
from fsspec.utils import tokenize

import torch 
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    torch_dtype=torch.bfloat16,
    device_map="auto"
)

messages = [
    {"role": "user", "content": "Hello, how are you?"},
    {"role": "user", "content": "mujhe enmbedding ke baare me btao!"},
    {"role": "assistant", "content": "How can i assist you sir"},
    {"role": "user", "content": "Explain about olama"}
]

prompt = pipe.tokenizer.apply_chat_template(messages,tokenize=False,add_generation_prompt=True)

output = pipe(prompt,max_new_tokens=20,temperature=0.7, do_sample=True)
print(output[0]["generated_text"])
```

```python
from langchain_huggingface import HuggingFacePipeline, ChatHuggingFace
from transformers import pipeline
import torch

# Pipeline banao (same as above)
hf_pipe = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    torch_dtype=torch.bfloat16,
    device_map="auto",
    max_new_tokens=300,
    temperature=0.7
)

llm = HuggingFacePipeline(pipeline=hf_pipe)

chat_model = ChatHuggingFace(llm=llm)

# Ab messages list do (ye important hai!)
from langchain_core.messages import HumanMessage, SystemMessage

response = chat_model.invoke([
    SystemMessage(content="Tu ek friendly pirate hai, har jawab pirate style mein de."),
    HumanMessage(content="Olama kya hai aur kyun multiple nahi support karta?")
])

print(response.content)
```
