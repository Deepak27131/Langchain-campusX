```python
import random 

class NakliLLM:

    def __init__(self):
        print("LLM Created")

    def predict(self,promptt):

        response_list = [
            'Delhi is the capital of india',
            'IPL is a cricket league',
            'AI stand for Artificial Intelligence'
        ]

        return {'response':random.choice(response_list)}
```

```python
llm = NakliLLM()

llm.predict("What is AI?")
```

```python
class NakliPromptTemplate:

    def __init__(self,template,input_variables):
        self.template = template
        self.input_variables = input_variables

    def format(self,input_variables):
        return self.template.format(**input_variables)
```

```python
template = NakliPromptTemplate(
    template='Write a {length} poem about {topic}?',
    input_variables=['topic','length']
)

print(template.format(input_variables={'topic':'India','length':'short'}))
```

```python
prompt = template.format(input_variables={'topic':'India','length':'short'})

```

```python
llm = NakliLLM()
```

```python
llm.predict(prompt)
```

```python
class NakliLLMChain:

    def __init__(self,llm,prompt):
        self.llm = llm
        self.prompt = prompt

    def run(self,input_dict):
        final_prompt = self.prompt.format(input_dict)
        result = self.llm.predict(final_prompt)
        return result['response']
        
```

```python
template = NakliPromptTemplate(
    template='Write a {length} poem about {topic}?',
    input_variables=['topic','length']
)
```

```python
llm = NakliLLM()
```

```python
chain = NakliLLMChain(llm,template)

print(chain.run({'topic':'India','length':'short'}))
```
