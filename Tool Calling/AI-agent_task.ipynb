```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langgraph.prebuilt import create_react_agent
import langchainhub as hub
```

```python
# step 2 :pull the react prompt from langchain
# prompt = hub.pull("hwchase17/react") # pull the standard React agent propmt
```

```python
prompt = "You are a helpful assistant. Use the search tool to find current info."
```

```python
# step 3 : Create the react agent manually with the pulled prompt
agent = create_react_agent(
    llm,
    tools=[search,get_weather_data],
    prompt = prompt
)
```

```python
response = agent.invoke({
    "messages": [("user", "3 ways to reach goa from delhi")]
})

# To print the answer:
print(response["messages"][-1].content)
```
